## 1. 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?

```python
max_temp = -999     # 최고 기온 값
max_date= ''         # 최고 기온 날짜

for row in arr:
    if row[-1] == '':
        row[-1] = -999       # 빈 문자열 처리
    row[-1] = float(row[-1]) # 최고 기온을 실수로 표현
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
        print(max_date, max_temp)
```

```python
print("서울의 최고 기온이 가장 높았던 날은 {}였고, {}도였다.".format(max_date, max_temp))
```

```python
print(np.max(arr[:, 4]))
```

## 2. 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?

```python
dif_temp = -999 # 일교차
dif_date = '' # 일교차

for row in arr:
    if row[2] == '':
        row[2] = 999       # 빈 문자열 처리
    row[2] = float(row[2]) 
    if (row[-1] - row[2]) >= dif_temp:
        if row[-1] == -999 or row[2] == 999:
            pass
        else:
            dif_temp = row[-1] - row[2]
            dif_date = row[0]
            print(dif_date, dif_temp)
```

```python
print("일교차가 가장 큰 날짜는 {}였다.".format(dif_date))
```
