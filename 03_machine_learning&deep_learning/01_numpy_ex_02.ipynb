```python
import numpy as np

ratings = np.loadtxt("./data/ratings.csv", dtype=np.object, 
                 delimiter=',', skiprows=1)
print(ratings)
# userId,movieId,rating,timestamp
```

```python
for row in ratings:
    row[0] = int(row[0])   # userId 
    row[1] = int(row[1])   # movieId
    row[2] = float(row[2]) # rating

print(ratings)
```

## 1. 사용자가 평가한 모든 영화의 전체 평균 평점

```python
print(np.mean(ratings[:, 2]))
```

## 2. 각 사용자별 평균 평점

```python
user_d = {}
movie_d = {}

for row in ratings:
    # 사용자별 평균 평점
    if row[0] in user_d.keys():
        user_d[row[0]][0] += row[2]
        user_d[row[0]][1] += 1
    else:
        user_d[row[0]] = [row[2], 1]
     
    # 영화별 평균 평점
    if row[1] in movie_d.keys():
        movie_d[row[1]][0] += row[2]
        movie_d[row[1]][1] += 1
    else:
        movie_d[row[1]] = [row[2], 1]
```

```python
# 사용자별 평균 평점
for k, v in user_d.items():
     print("{}번사용자: {}".format(k, v[0]/v[1]))
```

## 3. 각 영화별 평균 평점

```python
# 영화별 평균 평점
for k, v in movie_d.items():
     print("{}번영화: {}".format(k, v[0]/v[1]))
```
