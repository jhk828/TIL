# NumPy Exercise - 1 : 기온 분석
* 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?
* 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?
* 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?
* 가장 덥다고 알려진 대구보다 서울이 더 더운날이 가장 많은 연도는 언제인가요?

날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)

```python
import numpy as np

# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)
seoul = np.loadtxt("./data/seoul.csv", dtype=np.object, 
                 delimiter=',', skiprows=1)
daegu = np.loadtxt("./data/daegu.csv", dtype=np.object, 
                 delimiter=',', skiprows=1, encoding="utf-8")
```

```python
# 서울
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)
for row in seoul:
    if row[2] == '': # 평균기온
        row[2] = -999      
    else:
        row[2] = float(row[2]) 
    if row[3] == '': # 최저기온
        row[3] = -999      
    else:
        row[3] = float(row[3]) 
    if row[4] == '': # 최고기온
        row[4] = 999      
    else:
        row[4] = float(row[4]) 
print(seoul)
```

## 1. 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?

```python
max_temp = -999     # 최고 기온 값
max_date = ''       # 최고 기온 날짜

# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)
for row in seoul:
    if row[4] != 999 and row[4] >= max_temp:
        max_temp = row[4] 
        max_date = row[0]
#         print(max_date, max_temp)
        
print(max_date, max_temp)
```

### for문 안쓰고 풀기 (예시 답안)

- 데이터 분석을 진행할 때 데이터의 전처리가 진행되어야 한다.
- 결측치 처리와 이상치 처리를 해야 한다.
- **결측치 처리**
    - raw 데이터는 문제의 소지가 있을 수 있다.
    - 데이터가 없는 경우를 의미
    - 삭제/ 평균,최고, 최저 등으로 값 대체
- **이상치 처리**
    - 데이터는 존재하는데 정상범위가 아닌 데이터
    - 삭제하거나 대체

```python
import numpy as np

seoul2 = np.loadtxt("./data/seoul.csv", dtype=np.object, delimiter=',', skiprows=1)
print(seoul2)
```

```python
print(seoul2.T)
```

```python
# 최고 기온 
# 최고온도를 알아내서 boolean indexing으로 날짜를 알아낼 수 있다.
high_temp_seoul = seoul2.T[4]
print(high_temp_seoul)
# ['20.7' '22' '21.3' ... '27.5' '27.3' '24.4']

# 결측치 Boolean Mask
print(high_temp_seoul == '')
print((high_temp_seoul == '').sum())  # 758개
```

```python
# 결측치 삭제
# boolean indexing을 이용해서 ''가 아닌 최고기온만 뽑아낸다.
high_temp_seoul[high_temp_seoul != '']
# array(['20.7', '22', '21.3', ..., '27.5', '27.3', '24.4'], dtype=object)

high_temp = high_temp_seoul[high_temp_seoul != ''].astype(np.float64).max() # 39.6
```

```python
# 최고 기온을 알아냈으니, 최고 기온의 날짜 알아내기

# 최고기온 ndarray 안에서, 위에서 구한 최고 기온에 대한 boolean mask 만들기
print(high_temp_seoul == str(high_temp))  # 최대값만 True고 나머지는 False
print(seoul2.T[0][high_temp_seoul == str(high_temp)]) # 중복처리도 자동으로 됨
```

## 2. 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?

```python
dif_temp = -999 # 일교차
dif_date = '' # 일교차
# 날짜,지점,평균기온(℃),최저기온(℃),최고기온(℃)
for row in seoul:
    if row[3] != -999 and row[4] != 999 and (row[4]-row[3]) >= dif_temp:
        dif_temp = row[4]-row[3]
        dif_date = row[0]
#         print(dif_date, dif_temp)
        
print(dif_date, dif_temp)
```

### for문 안쓰고 풀기

```python
seoul2 = np.loadtxt("./data/seoul.csv", dtype=np.object, 
                 delimiter=',', skiprows=1)

print(seoul2)
```

```python
print(seoul2[:,3]) # ['7.9' '7.9' '13.1' ... '19.9' '19.2' '19.5']

empty_low_idx = np.where(seoul2[:,3]=='')
seoul2 = np.delete(seoul2, empty_low_idx, axis=0)  # 최저기온 "" 행 제거
empty_high_idx = np.where(seoul2[:,4]=='')
seoul2 = np.delete(seoul2, empty_high_idx, axis=0)  # 최고기온 "" 행 제거
```

```python
a = (seoul2[:,-1].astype(np.float64) - seoul2[:,-2].astype(np.float64)).reshape(-1, 1)
print(a)
# [[12.8]
#  [14.1]
#  [ 8.2]
#  ...
#  [ 7.6]
#  [ 8.1]
#  [ 4.9]]
```

```python
result = np.concatenate((seoul2,(seoul2[:,-1].astype(np.float64) - seoul2[:,-2].astype(np.float64)).reshape(-1, 1)),axis=1) 
print(result)
```

```python
max_high = result[:,-1].max() # 최대 온도
print(max_high)
print(result[:,0][result[:,-1] == max_high])
```

## 3. 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?

```python
day = {}
for row in seoul:
    if row[0][5:] in day.keys():
        day[row[0][5:]][0] += row[4]-row[3]
        day[row[0][5:]][1] += 1
    else:
        day[row[0][5:]] = [row[4]-row[3], 1]
```

```python
print(day)
```
